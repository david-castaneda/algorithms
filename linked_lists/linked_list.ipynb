```python
class SinglyNode:
    
    def __init__(self, val, next = None):
        self.val = val
        self.next = next
    
    def __str__(self):
        return str(self.val)
    
Head = SinglyNode(1)
A = SinglyNode(2)
B = SinglyNode(3)
C = SinglyNode(4)

Head.next = A
A.next = B
B.next = C

print(Head)
```

```python
# Time: O(n), Space: O(n)
def printReverseOrder(node):
    if node is None:
        return
    
    printReverseOrder(node.next) # recursively until the end of the list
    
    print(node)

printReverseOrder(Head)
```
