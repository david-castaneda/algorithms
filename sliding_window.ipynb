# Sliding Window

Useful when looking at substrings or subarrays that fit some criteria.

Types of sliding window:
1. Variable-Length
2. Fixed-Length

# Variable-Length

When can we apply sliding window?
1. We're looking for a min/max length substring/subarray
2. The is some condition that adjust based on the window size dynamically updating

## Variable Length Sliding Window - Pattern

```python
Function VariableLengthSlidingWindow(nums, ...):
    Setup tracking variables e.b MaxW, NumZeros
    L = 0
    For R in Range(N):
        Add Nums[R] to Window
        
        While Window Is Not Valid:
            Remove Nums[L] From Window
            L += 1
    
        W = (R-L+1)
        Check for new Max/Min
    
    Return MaxW
```

````python
# Example 1 - Max Consecutive Ones III
# Time: O(n)
# Space: O(1)
input  = [1,1,1,0,0,0,1,1,1,1]
k = 2

l = 0
max_w = 0
num_zeros = 0

for r in range(len(input)):
    if input[r] == 0:
        num_zeros += 1
    
    while num_zeros > k:
        if input[l] == 0:
            num_zeros -= 1
        l += 1
        
    max_w = max(max_w, r - l + 1)
    
print(max_w)
````

````python
# Example 2 - Longest Substring
# Time: O(n)
# Space: O(n)
input = "abcbcdef"

l = 0
max_w = 0
seen = set()

for r in range(len(input)):
    while input[r] in seen:
        seen.remove(input[l])
        l += 1
    
    new_w = (r - l + 1)
    max_w = max(max_w, new_w)
    
    seen.add(input[r])
    
print(max_w)
````
