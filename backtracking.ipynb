# Recursive Backtracking

Useful for subset problems where we need to generate ALL solutions to a problem.

It's exaustive and looks at everything.

- Time: O(2^n)
- Space: O(n)

At each step
- Pick a value, recurse
- Don't pick a value, recurse

```python
# Time: O(2^n)
# Space: O(n)
def subsets_of_array(arr):
    answer = []
    solution = []
    
    def helper(i):
        if i == len(arr):
            answer.append(solution.copy()) # important to make a copy
            return

        # 1 - exclude path
        helper(i + 1)
        
        # 2 - include path
        solution.append(arr[i])
        helper(i + 1)
        solution.pop() # revert
        
    helper(0)
    
    return answer

subsets_of_array([1,2,3])
```
