# Recursive Backtracking

Useful for subset problems where we need to generate ALL solutions to a problem.

It's exaustive and looks at everything.

- Time: O(2^n)
- Space: O(n)

At each step
- Pick a value, recurse
- Don't pick a value, recurse

```python
# Time: O(2^n)
# Space: O(n)
def subsets_of_array(arr):
    answer = []
    solution = []
    
    def backtrack(i):
        if i == len(arr):
            answer.append(solution.copy()) # important to make a copy
            return

        # 1 - exclude path
        backtrack(i + 1)
        
        # 2 - include path
        solution.append(arr[i])
        backtrack(i + 1)
        solution.pop() # revert
        
    backtrack(0)
    
    return answer

subsets_of_array([1,2,3])
```

# Permutations

- Number of branches is based on N - indicates a loop

```python
# Time: O(n!)
# Space: O(n)
def permutations_of_array(arr):
    answer = []
    solution = []
    
    def backtrack():
        if len(solution) == len(arr):
            answer.append(solution.copy())
            return
        
        for num in arr:
            if num not in solution: # check item not in solution
                solution.append(num)
                backtrack()
                solution.pop()
            
    backtrack()
    
    return answer

permutations_of_array([1,2,3])
```
