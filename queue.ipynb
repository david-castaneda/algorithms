#### Queue (FIFO data structure)

Best practice: Use `collections.deque`

- insert (enqueue): O(1)
- peek: O(1)
- pop: O(1)

```python
from collections import deque

q = deque()
q.append(1) # enqueue right
q[0] # peek front (left)
q.popleft() # dequeue left

# Check if queue is empty
if q:
    print("Not empty")
else:
    print("Empty")
```
