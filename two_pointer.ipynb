# Two-pointer

Technique that uses two pointers that start at opposite ends of an array and gradually move towards each other.

```python
nums = [1,2,3,4,5,6,7,8]
left = 0 # left pointer
right = len(nums) - 1 # right pointer

while left < right:
    print(nums[left], nums[right])
    
    left += 1
    right -= 1
```

```python
# Example - Given a sorted array of integers nums, determine if there exists a pair of numbers that sum to a given target.
def two_sum(nums, target):
    left = 0
    right = len(nums) - 1
    
    while left < right:
        cur_sum = nums[left] + nums[right]
        
        if cur_sum == target:
            return True
        
        if cur_sum < target:
            left += 1
        else:
            right -= 1
    
    return False

two_sum([1,2,3,4,5], 4)
```

## Summary 

- Two pointer technique leverages the fact that the input array is sorted
- Significantly improve time complexity from O(n^2) to O(n)

## When to use?

Consider using the two-pointer technique for questions that involve searching for a pair (or more) of items in an array that meet a certain criteria.

- Finding a pair of items that sum to a given target in an array
- Finding a triplet of items that sum to 0 in a given array
- Finding the maximum amount of water that can be held between two array items representing wall heights
