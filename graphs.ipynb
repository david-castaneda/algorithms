# Graphs

- 2 main types: Undirected and Directed
- 4 common ways to store a graph: Edge List, Adjacency Matrices, Adjacency Lists, and Class-based

Terminology:
- Nodes / Vertices contain values 
- Edges are connections between Nodes
- Cyclic is when a graph has cycles
- Acyclic is no cycles
- InDegree is the number of connection that point in to it
- OutDegree is the number of connections that point out from it
- Neighbors are the vertices you can directly reach from a given node

## Mathematically

A Graph is a set of Vertices (Nodes)

V = {Colombia, US, Japan}
E = { (US, Canada), (US, Japan), (Japan, Colombia) }

The above is an example of a directed graph. E = { (From, To) }
