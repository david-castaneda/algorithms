#### Stack (LIFO data structure)

- insert (push): \*O(1) amortized
- peek: O(1)
- pop: O(1)

```python
stack = [] # list used as a stack
stack.append(1) # push to end
stack[-1] # peek
stack.pop() # prop from end

# Check if stack is empty
if stack:
    print("Not empty")
else:
    print("Empty")
```
