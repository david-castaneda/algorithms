# Trees

- 1 or more children

# Binary Trees

- 0, 1 or 2 children (left and right)

# Characterstics of trees
- Perfect Tree - All rows are perfectly full
- Complete - Every row is perfectly full except the last row. Must be full left-to-right
- Height - Maximum depth from **root to leaf**

# Traversing a tree
- Depth First Search(DSF) - Most common. Uses a Stack data structure.
    - Time: O(n)
    - Space: O(H) - height of the tree = call stack. Worst case H = N for skewed trees.
    - Three flavors of DSF:
        1. PreOrder - Root, Left, Right
        2. InOrder - Left, Root, Right
        3. PostOder - Left, Right, Root
- Breadth First Search (BFS) - Level-Order Traversal. Uses a Queue data structure.
    - Time: O(n)
    - Space: O(n) - Last level on the tree is n/2.

```python
# Template
class TreeNode:

    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
    
    def __str__(self):
        return str(self.val)
```

```python
#       1
#    2     3
#  4   5  6
Root = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(6)

Root.left = B
Root.right = C
B.left = D
B.right = E
C.left = F

print(Root)
```

# Depth First Search (DFS)

```python
# Recursive DFS
def dfs(node):
    # base case
    if node is None:
        return
    
    print(node.val) # process current node
    
    # recursive calls - prioritize depth
    dfs(node.left)
    dfs(node.right)
    
dfs(Root)
```

```python
# Iterative DFS - Forces Pre order traversal
def iterativeDFS(node):
    stack = [node]
    while stack:
        cur = stack.pop()
        
        print(cur.val) # process current node
        
        # order matters - right then left - force stack to go left first
        if cur.right:
            stack.append(cur.right)
        if cur.left:
            stack.append(cur.left)
            
iterativeDFS(Root) 
```

```python
# Pre order traveral: Root, Left, Right
def preOrderDFS(node):
    if node is None:
        return
    
    print(node.val) # process current node
    
    dfs(node.left)
    dfs(node.right)

preOrderDFS(Root)
```

```python
# In order traveral: Left, Root, Right
def inOrderDFS(node):
    if node is None:
        return
    
    dfs(node.left)
    
    print(node.val) # process current node
    
    dfs(node.right)

inOrderDFS(Root)
```

```python
# Post order traveral: Left, Right, Root
def postOrderDFS(node):
    if node is None:
        return
    
    dfs(node.left)
    dfs(node.right)
    
    print(node.val) # process current node

postOrderDFS(Root)
```

# Breadth First Search (BFS)

```python
from collections import deque

# BFS -  Level Order Traversal
def bfs(node):
    q = deque()
    q.append(node)
    
    while q:
        cur = q.popleft()
        
        print(cur.val) # process current node
        
        if cur.left:
            q.append(cur.left)
        if cur.right:
            q.append(cur.right)

bfs(Root)
```

```python
# Height of binary tree using DFS
def maxHeight(node):
    if not node:
        return 0

    left = maxHeight(node.left)
    right = maxHeight(node.right)
    
    return 1 + max(left, right)

maxHeight(Root)
```

```python
# Inverting a binary tree using DFS
def invertBinaryTree(node):
    if not node:
        return
    
    node.left, node.right = node.right, node.left
    
    invertBinaryTree(node.left)
    invertBinaryTree(node.right)

invertBinaryTree(Root)
print(Root)
```
