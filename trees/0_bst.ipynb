# Binary Search Tree (BST)

Special kind of tree where:
- Every node to the left is less than current node
- Every node to the right is greater than current node

```python
# Template
class TreeNode:

    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
    
    def __str__(self):
        return str(self.val)
```

```python
#         4
#      2     6
#   1   3  5   7
Root = TreeNode(4)

A = TreeNode(2, TreeNode(1), TreeNode(3))
B = TreeNode(6, TreeNode(5), TreeNode(7))

Root.left = A
Root.right = B

print(Root)
```

```python
# Look up / Find
# Time: O(H) - For balanced BTS height is roughly O(logn)
# Space: O(H) - For balanced BTS height is roughly O(logn)
def search_bst(node, target):
    if not node:
        return False
    
    if node.val == target:
        return True
    
    if target < node.val:
        return search_bst(node.left, target)
    else:
        return search_bst(node.right, target)
    
search_bst(Root, 4)
```

## Balanced BST

For balanced binary search trees the height will be approximately O(logn) where n is the number of nodes in the tree.

For BST problems we assume the tree is balanced.

- Look up in balanced BST - Time: O(logn), Space O(logn)
- Look up in unbalanced BST - Time: O(n), Space O(n)
- InOrder DFS - Time O(n)

```python
# Time: O(n)
# Space: O(n)
# Intuition: Use a min/max range based on which side you are traversing.
def valid_bst(node, min, max):
    stack = [(node, min, max)]
    
    while stack:
        node, min, max = stack.pop()
    
        if node.val <= min or node.val >= max:
            return False
            
        if node.left:
            stack.append((node.left, min, node.val))
        if node.right:
            stack.append((node.right, node.val, max))
        
    return True

valid_bst(Root, float('-inf'), float('inf'))
```
