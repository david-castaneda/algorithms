#### Heaps / Priority Queues

- Min heap - binary tree where every node is smaller than all the values in its subtree
- Max heap - binary tree where every node is larger than all the values in its subtree

Binary trees can be represented using an array.

- peek: O(1)
- pop: O(logn)
- push: O(logn)
- heapify: O(n)

```python
# Build Min Heap (Heapify)
# Time: O(n), Space: O(1)

A = [4,7,1,3,2,5,6]

import heapq
heapq.heapify(A)

print(A)
```

```python
# Peak at Min: Time O(1)

A[0]
```

```python
# Heap Push (Insert Element)
# Time: O(logn)

heapq.heappush(A, 0)

print(A)
```

```python
# Heap Pop (Extract min)
# Time: O(logn)

minn = heapq.heappop(A)

print(A, minn)
```

```python
 # Max Heapify 
B = [-4,3,1,0,2,5,10,8,12,9]

# Negate every value 
for i in range(len(B)):
    B[i] = -B[i]

heapq.heapify(B)

print(B)
```

```python
# Extracting largest value
largest = -heapq.heappop(B) # negate the value

print(largest)
```

```python
# Insert into max heap
heapq.heappush(B, -7)

print(B)
```

```python
# Putting tuples on the heap (pt. 1)

C = [1,2,2,1,3,3,2,1,4,4,4,5]

from collections import Counter

counter = Counter(C)

print(counter)
```

```python
# Putting tuples on the heap (pt. 2)

heap = []

for k ,v in counter.items():
    heapq.heappush(heap, (v, k))
    
print(heap)
```
